Importing The Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
loan_dataset = pd.read_csv('dataset.xls')
```

```python
# printing the first 5 rows of the DataFrame
loan_dataset.head()
```

Statistical Measures for this Dataset

```python
loan_dataset.describe()
```

```python
loan_dataset.isnull().sum()
```

```python
#dropping the missing values
loan_dataset = loan_dataset.dropna()
```

```python
loan_dataset.isnull().sum()
```

```python
# label encoding
loan_dataset.replace({"Loan Status": {'N':0,'Y':1}}, inplace = True)
```

```python
loan_dataset.head()
```

```python
loan_dataset['Dependents'].value_counts()
```

```python
#replacing the value of 3+ to 4
loan_dataset= loan_dataset.replace(to_replace='3+', value=4)
```

```python
#dependent values
loan_dataset['Dependents'].value_counts()
```

Data Visualization

```python
#education and loan status
sns.countplot(x='Education', hue='Loan_Status', data=loan_dataset)
```

Marital Status and Loan Status

```python
sns.countplot(x='Married', hue='Loan_Status', data=loan_dataset)
```

Convert Categorical Columns to Numerical Values

```python
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace = True)
```

```python
loan_dataset.head()
```

Seperating the Data and Label

```python
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'], axis=1)
```

```python
Y = loan_dataset['Loan_Status']
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, stratify = Y, random_state= 42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model

Support Vector Machine Model

```python
classifier = svm.SVC(kernel= 'linear')
```

Training the SVM model

```python
classifier.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score on Training Data

```python
X_training_prediction = classifier.predict(X_train)
training_data_accuarcy = accuracy_score(X_training_prediction, Y_train)
```

```python
print("Accuracy on training data:",training_data_accuarcy)
```

```python
X_testing_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_testing_prediction, Y_test)
```

```python
print("Accuracy on testing data:",training_data_accuarcy)
```
